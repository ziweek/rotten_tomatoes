```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_names = os.listdir(path='csv')
movie_names = [csv_name.replace('.csv', '') for csv_name in csv_names]
if '.DS_Store' in movie_names:
    movie_names.remove('.DS_Store')
movie_names.sort(reverse=False)
movie_names
```

```python
f1 = plt.figure(figsize=(30,45))
for i, movie_name in enumerate(movie_names):
    df = pd.read_csv('csv/' + movie_name + '.csv', index_col='Title')
    for j, e in enumerate(df['RatingTomatometer'].astype(str)):
        df['RatingTomatometer'].iloc[j] = e.replace('%', '')
    for k, e in enumerate(df['Box Office (Gross USA)'].astype(str)):
        df['Box Office (Gross USA)'].iloc[k] = e.replace('$', '').replace('M', '').replace('K','')
    ax1 = plt.subplot(4, 5, i+1)
    sns.regplot(x=df['RatingTomatometer'].astype(float), y=df['Box Office (Gross USA)'].astype(float))
    # plt.scatter(df['RatingTomatometer'].astype(float), df['Box Office (Gross USA)'].astype(float),alpha=.3)
    plt.title(movie_name)
    plt.xlabel('TOMATOMETER')
    plt.ylabel('BOX-OFFICE')
plt.show()
# plt.savefig('default.png')
```
